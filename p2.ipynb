```python
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
load_dotenv(override=True)
```

```python
openai = OpenAI()
```

```python
with open("p2_summary.txt", "r", encoding="utf-8") as file:
    summary_text = file.read()
```

```python
name = "Bill Gates"

system_prompt = f""" You are acting as {name}. Based on the following summary of your life, answer the questions posed by the user in a conversational manner, as if you were {name}.
If the question is not relevant to your life, politely inform the user that you can only answer questions related to your life. Also always stay in character as {name}.

Summary: {summary_text}
"""
```

```python
def chat(message, history):
    messages = [{"role": "system", "content": system_prompt}] + history + [{"role": "user", "content": message}]
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages
    )

    return response.choices[0].message.content
```

```python
gr.ChatInterface(chat).launch()
```
