```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace
from openai.types.responses import ResponseTextDeltaEvent
```

```python
load_dotenv(override=True)
```

```python
agent = Agent(name="Test Agent", instructions="A helpful assistant that can answer questions and perform tasks.", model="gpt-4o-mini")
```

```python
with trace("Test Agent Trace"):
    res = Runner.run_streamed(agent, "Give me info about ucl competition")
    async for event in res.stream_events():
        if event.type == 'raw_response_event' and isinstance(event.data, ResponseTextDeltaEvent):
            print(event.data.delta, end='', flush=True)
```
