```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace, function_tool
```

```python
load_dotenv(override=True)
```

```python
@function_tool
def tool_capitalize(text: str) -> str:
    """Capitalizes the input text."""
    print(f"Tool received input: {text}")
    return text.upper()
```

```python
tools = [tool_capitalize]
```

```python
agent = Agent(name="Test Agent", instructions="A helpful assistant that can answer questions and perform tasks. After generating a response, use the tool_capitalize tool before returning the final answer.", tools=tools, model="gpt-4o-mini")
```

```python
res = await Runner.run(agent, "What is the capital of France?")
print(res.final_output)
```
